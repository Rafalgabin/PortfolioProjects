```python
#As per graphs we can notice some pattern, hihghest attrition is around age of 30 and slowly goes down with higher age but there is small spike at age of 40 and 50, after it its slowly rising.
#I would assume that with bigger distance to work from home attrition would be higher but in this data sample its not true. 
#The higher monthly salary is the less atrrition but shortly after promotion attrition is the highest in terms of years since last promotion.
#The longer someone is working in this company and the longer on the same role the attrition goes down with small spike after 10 years of work.
#Workers with lower job level, none or low salary hike and none or low stock option level are more likely to quit.
#More traveling workers may quit more often than those that not travel in case of business travels.
#Most likely to quit their job are workers with Human Resources, Technical Degree or Marketing Education Field.
#Sales Representatives are most likely to quit compared to other Job Roles.
#Workers which are single are more likely to quit, same goes with those who work overtimes.
```

```python
df['Education'] = df['Education'].astype(int)

# Mapping numeric values to corresponding strings
education_mapping = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}

# Apply the mapping to the Education column
df['Education'] = df['Education'].map(education_mapping)

df
```

```python
df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis="columns", inplace=True)
```
